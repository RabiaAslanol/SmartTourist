```python
import os
import uuid
from icrawler.builtin import GoogleImageCrawler

# Kategoriler ve arama anahtar kelimeleri
kategoriler = {
    'tarihi': ['tarihi cami', 'kale', 'tarihi eser', 'saray'],
    'dogal': ['doğal göl', 'şelale', 'dağ manzarası', 'orman'],
    'muze': ['etnografya müzesi', 'arkeoloji müzesi', 'sanat müzesi'],
    'plaj': ['kumsal', 'deniz kenarı', 'plaj', 'beach', 'güneşlenme alanı'],
}

hedef_sayi = 300  # Her klasörde olması gereken hedef foto sayısı

# Her kategori için ayrı işlem
for kategori, anahtar_liste in kategoriler.items():
    klasor_yolu = os.path.join("dataset", kategori)
    os.makedirs(klasor_yolu, exist_ok=True)

    mevcut_resim_sayisi = len([
        f for f in os.listdir(klasor_yolu)
        if f.lower().endswith(('.jpg', '.jpeg', '.png'))
    ])

    kalan = hedef_sayi - mevcut_resim_sayisi

    if kalan > 0:
        print(f"\n📥 {kategori} klasörü için {kalan} yeni foto indirilecek.")
        for anahtar in anahtar_liste:
            if kalan <= 0:
                break

            indirilecek = min(kalan, 40)

            crawler = GoogleImageCrawler(
                storage={'root_dir': klasor_yolu}
            )

            # Geçici klasöre indir
            temp_dir = os.path.join(klasor_yolu, "temp")
            os.makedirs(temp_dir, exist_ok=True)
            crawler = GoogleImageCrawler(storage={'root_dir': temp_dir})
            crawler.crawl(keyword=anahtar, max_num=indirilecek)

            # Geçici klasörden eşsiz adlarla taşı
            for fname in os.listdir(temp_dir):
                if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
                    yeni_ad = f"{uuid.uuid4()}.jpg"
                    os.rename(os.path.join(temp_dir, fname), os.path.join(klasor_yolu, yeni_ad))

            os.rmdir(temp_dir)

            mevcut_resim_sayisi = len([
                f for f in os.listdir(klasor_yolu)
                if f.lower().endswith(('.jpg', '.jpeg', '.png'))
            ])
            kalan = hedef_sayi - mevcut_resim_sayisi

        print(f"✅ {kategori} tamamlandı: {mevcut_resim_sayisi} foto.")
    else:
        print(f"✔️ {kategori} zaten yeterli ({mevcut_resim_sayisi}) foto içeriyor.")
```
